**Demonstrate the data distributions using box, scatter plot, histogram, and bar chart on iris
dataset.**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#iris = sns.load_dataset('iris')
data = pd.read_csv("/content/drive/MyDrive/iris.csv")
```

```python
# Creating a grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Boxplot of Sepal Length
sns.boxplot(x="Flowers", y="sepallength", data=data, ax=axes[0, 0])
axes[0, 0].set_title('Boxplot of Sepal Length')

# Boxplot of Sepal Width
sns.boxplot(x="Flowers", y="sepalwidth", data=data, ax=axes[0, 1])
axes[0, 1].set_title('Boxplot of Sepal Width')

# Boxplot of Petal Length
sns.boxplot(x="Flowers", y="petallength", data=data, ax=axes[1, 0])
axes[1, 0].set_title('Boxplot of Petal Length')

# Boxplot of Petal Width
sns.boxplot(x="Flowers", y="petalwidth", data=data, ax=axes[1, 1])
axes[1, 1].set_title('Boxplot of Petal Width')

# Adjusting spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.5)

# Showing the plot
plt.show()
```

```python
# Scatterplot
sns.scatterplot(x="sepallength", y="sepalwidth", hue="species", data=data)
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.show()
```

```python
# Creating a single figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Drawing Histograms for each column
sns.histplot(data['sepallength'], ax=axes[0, 0], color="skyblue")
axes[0, 0].set_title('Sepal Length')

sns.histplot(data['sepalwidth'], ax=axes[0, 1], color="yellow")
axes[0, 1].set_title('Sepal Width')

sns.histplot(data['petallength'], ax=axes[1, 0], color="magenta")
axes[1, 0].set_title('Petal Length')

sns.histplot(data['petalwidth'], ax=axes[1, 1], color="lightgreen")
axes[1, 1].set_title('Petal Width')

# Adjusting layout
plt.tight_layout()

# Showing the plots
plt.show()
```

```python
# Bar chart
data['Flowers'].value_counts().plot(kind='bar')
plt.title('Bar Chart of Species')
plt.show()
```
