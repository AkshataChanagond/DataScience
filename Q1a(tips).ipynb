```python
import matplotlib.pyplot as plt

plt.hist(data['total_bill'], bins=20, edgecolor='black',color='teal')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

**Plot a histogram for the variable ‘TotalBill’ to check which range has the highest frequency.**

```python
day_counts = data['day'].value_counts()

# Create a bar chart using the matplotlib library
plt.bar(day_counts.index, day_counts.values)

# Add labels and a title to the bar chart
plt.xlabel('Day')
plt.ylabel('Frequency')
plt.title('Bar Chart of Day')
```

```python
max_day = day_counts.idxmax()
print('The category with the maximum count is:', max_day)
```
